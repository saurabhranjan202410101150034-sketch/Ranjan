```python
df = pd.read_csv('/content/drive/MyDrive/Online Retail Data Set.csv', encoding='latin1')
df
```

```python
df = df.dropna()
df = df.drop_duplicates()
print(df.isnull().sum())
print(df.duplicated().any())

df['CustomerID'] = df['CustomerID'].apply(lambda x: int(x))
df['CustomerID'] = df['CustomerID'].apply(lambda x: repr(x))
df.info()
```

```python
df.describe(include='all')
```

```python
df['Description'] = df['Description'].str.lower()
df['Country'] = df['Country'].str.lower()
df = df[df['Quantity'] > 0]
df = df[df['UnitPrice'] > 0]
df.reset_index(inplace=True)
print(df.isnull().sum())
print(df.duplicated().any())
```

```python
def remove_outlier(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR

    # Create arrays of Boolean values indicating the outlier rows
    idx = df[(df[col] >= upper) | (df[col] <= lower)].index
    df.drop(idx , inplace=True)
    # Removing the outliers
#     df.drop(index=upper_array, inplace=True)
#     df.drop(index=lower_array, inplace=True)

    # Removing the outliers
#     df.drop(index=upper_array, inplace=True)
#     df.drop(index=lower_array, inplace=True)
```

```python
remove_outlier(df, 'Quantity')
remove_outlier(df, 'UnitPrice')
```

```python
df.describe()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='mixed', dayfirst=True)
df['InvoiceDate'].dt.year.unique()
```

```python
df_time = df.copy()
df_time['date'] = df_time['InvoiceDate'].dt.normalize()
df_time['day'] = df_time['InvoiceDate'].dt.day_name()
```

```python
grouped_date = df_time.groupby("date").agg({"Quantity": np.sum, "InvoiceNo": pd.Series.nunique}).reset_index()
grouped_date
```

```python
grouped_date['NormQuantity'] = (grouped_date['Quantity'] - grouped_date['Quantity'].min()) / (grouped_date['Quantity'].max() - grouped_date['Quantity'].min())
grouped_date['NormInvoiceCount'] = (grouped_date['InvoiceNo'] - grouped_date['InvoiceNo'].min()) / (grouped_date['InvoiceNo'].max() - grouped_date['InvoiceNo'].min())
ax = plt.gca()
grouped_date.plot( x = 'date' , y = 'NormQuantity', ax = ax )
grouped_date.plot( x = 'date' , y = 'NormInvoiceCount' , ax = ax )
```

```python
grouped_day = df_time.groupby("day").agg({"Quantity": np.sum, "InvoiceNo": pd.Series.nunique}).reset_index()
grouped_day
```

```python
invtotal = df.copy()
invtotal['total'] = df['Quantity']*df['UnitPrice']
```

```python
custtotal = invtotal.groupby('CustomerID').agg({"InvoiceNo": pd.Series.nunique, "Quantity": np.mean, "UnitPrice": np.mean, "total": np.sum})
x_arr = np.array(custtotal)
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_arr)
kmeans = KMeans(n_clusters = 3, random_state=0)
kmeans.fit(x_scaled)
kmeans.cluster_centers_
```

```python
custtotal['cluster'] = kmeans.labels_
ax = plt.scatter(x_scaled[:,1], x_scaled[:,2], s = 50, c = custtotal['cluster'], marker = 'o', alpha = 1, cmap='Set2')
centers = kmeans.cluster_centers_
plt.scatter(centers[:,1], centers[:,2], s=200, alpha=1 , marker='^')
plt.colorbar(ax)
plt.xlabel("Quantity")
plt.ylabel("UnitPrice")
plt.show()
```
